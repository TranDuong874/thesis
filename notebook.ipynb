```python
import open3d as o3d
import time

INPUT_PCD = "poses_points.pcd"
OUTPUT_PLY = "poses_points.ply"

print("Loading downsampled PCD...")
t0 = time.time()
pcd = o3d.io.read_point_cloud(INPUT_PCD)
t1 = time.time()

print(f"Loaded in {t1 - t0:.2f}s")
print(f"Points: {len(pcd.points):,}")

print("Saving as PLY...")
o3d.io.write_point_cloud(
    OUTPUT_PLY,
    pcd,
    write_ascii=False   # binary PLY = smaller + faster in Blender
)

print("Done.")
```
